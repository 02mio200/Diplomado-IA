## la columna de Paid Fees es un objeto, se convertira a float

```python

# Remover el símbolo de dólar y convertir la columna 'Paid Fees' a numérico
df['Paid Fees'] = df['Paid Fees'].replace('[\$,]', '', regex=True)

# Limpiar la columna 'Paid Fees' de espacios adicionales y convertir a numérico
df['Paid Fees'] = df['Paid Fees'].str.replace('USD ', '').str.replace(',', '').astype(float)
```
