# Tratamiento de datos

```python
import numpy as np
from sklearn.impute import SimpleImputer
from sklearn.impute import KNNImputer

impute_numeric =SimpleImputer(strategy='mean')

numeric_colums = df.select_dtypes(include='number').columns

df[numeric_colums] = impute_numeric.fit_transform(df[numeric_colums])
```

```python
impute_categorico = SimpleImputer(strategy='most_frequent')

categorical_colums = df.select_dtypes(include='object').columns
df[categorical_colums] = impute_categorico.fit_transform(df[categorical_colums])
```

```python
df.isnull().sum()
```

# Tratar con datos nominales y ordinales

```python
for name in df.columns.to_list():
    print(f'{name} : {df[name].unique()}')
    print('\n')
```

```python
# tratar con variables ordinales

Attrition_labesl = {'Yes':1,'No':0}

#mapear
df['Attrition'] = df['Attrition'].map(Attrition_labesl)
```

```python
# variables nominales

df = pd.get_dummies(df, columns=['BusinessTravel','EducationField','Department','MaritalStatus','JobLevel','OverTime','Gender','JobRole'])
```

# DETECCION DE VALORES ATIPICOS

```python
#%pip install seaborn
```

```python

import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
# calculando datos no atipicos
Q1 = df['Age'].quantile(0.25)
Q3 = df['Age'].quantile(0.75)

IQR = Q3 - Q1
limite_inferior = Q1 - 1.5 * IQR

limite_superior = Q3 + 1.5 * IQR

print(limite_inferior,limite_superior)
```

## Simulacion de datos atipicos

```python
df_test = df.copy()

indice = np.random.choice(df_test.index, 2, replace=False)

df_test.loc[indice[0],'Age'] += 3 * IQR
df_test.loc[indice[1],'Age'] += 3 * IQR
```

```python
#plt.hist(df['Age'],bins=30)

sns.displot(x = df['Age'])
```

```python
df_test.boxplot(column='Age')
```

```python
# calculando el rango de los datos atipicos atipidos
Q1 = df_test['Age'].quantile(0.25)
Q3 = df_test['Age'].quantile(0.75)

IQR = Q3 - Q1
limite_inferior = Q1 - 1.5 * IQR

limite_inferior

limite_superior = Q3 + 1.5 * IQR

limite_superior

print(limite_inferior,limite_superior)
```
