```python
experiencia = [1,2,4,7,8,10,12,15,20]
salario_data_scienlsts = [20000,22000,21000,23000,21000,20000,24000,21000,20000]
salario_ing_sotfware = [20000,20000,29000,20008,20001,26000,25000,20003,24000]
```

```python
plt.plot(experiencia,salario_data_scienlsts)
plt.plot(experiencia,salario_ing_sotfware)
plt.title('salario')
plt.xlabel('experiencia')
plt.ylabel('ganancia')
plt.legend(['data','ing'])
```

```python
plt.plot(experiencia,salario_data_scienlsts, color = 'b', label = 'data')
plt.plot(experiencia,salario_ing_sotfware,  color = 'k', label = 'ing')
plt.title('salario')
plt.xlabel('experiencia')
plt.ylabel('ganancia')
plt.legend()
```

```python
plt.plot(experiencia,salario_data_scienlsts, color = 'b', linestyle = '-.' , marker = 'o')
plt.plot(experiencia,salario_ing_sotfware,  color = 'k', linestyle = ':', marker = '*' )
plt.title('salario')
plt.xlabel('experiencia')
plt.ylabel('ganancia')
plt.legend(['data','ing'])
plt.grid(color = 'r', linestyle= '--')
```

## agregando numpy para datos aleatorios

```python
import numpy as np 
```

```python
x = np.random.rand(100)
y = np.random.rand(100)

```

## grafico dispersion

```python
plt.scatter(x,y)
plt.title('Grafico de dispersion')
```

## histograma

```python
datos = np.random.randn(100000)

plt.hist(datos,bins=10000)
plt.title('valor')
plt.xlabel('frecuecia')
```

## Cajas y bijotes

```python
datos = [np.random.normal(0,std,100) for std in range(1,4)]

plt.boxplot(datos,vert=True)
```

## serie de tiempo

```python
import pandas as pd

data_range = pd.date_range(start='2023-01-01',end = '2023-12-13',freq='D')
precios = np.cumsum(np.random.randn(len(data_range))) + 100
```

```python
data = pd.DataFrame({'Date':data_range, 'Precios':precios})
```

```python
plt.plot(data['Date'],data['Precios'],  label = 'Precios')
plt.show()
```

```python
plt.plot(data_range,precios)
```
